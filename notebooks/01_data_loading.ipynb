```python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# STEP 1: Path of one patient folder
# Change this path according to your folder location

patient_path = "../dataset/BraTs2020/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001/"

# STEP 2: File paths for FLAIR image and segmentation mask

flair_path = os.path.join(patient_path, "BraTS20_Training_001_flair.nii")
seg_path = os.path.join(patient_path, "BraTS20_Training_001_seg.nii")

# STEP 3: Load .nii files using nibabel

flair = nib.load(flair_path)
seg = nib.load(seg_path)

# STEP 4: Convert .nii files into NumPy arrays

flair_data = flair.get_fdata()
seg_data = seg.get_fdata()

# STEP 5: Print shape of both files

print("FLAIR MRI Shape:", flair_data.shape)
print("Segmentation Mask Shape:", seg_data.shape)

# Expected output example:
# (240, 240, 155)

# STEP 6: Choose one slice (middle slices are usually best)

slice_num = 75

# STEP 7: Display MRI slice and segmentation mask side by side

plt.figure(figsize=(10, 5))

# Left → MRI image
plt.subplot(1, 2, 1)
plt.imshow(flair_data[:, :, slice_num], cmap="gray")
plt.title("FLAIR MRI Slice")
plt.axis("off")

# Right → Tumor mask
plt.subplot(1, 2, 2)
plt.imshow(seg_data[:, :, slice_num], cmap="gray")
plt.title("Segmentation Mask")
plt.axis("off")

plt.tight_layout()
plt.show()
```
